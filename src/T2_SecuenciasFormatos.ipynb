## <span style="color:yellow"> Ejercicio_2</span>
    
1. Teniendo como input un archivo fasta, crea un diccionario de todas las secuencias en tu atchivo FASTA.

2. Encuentra todos los codones en tu primer marco de lectura. 

3. Imprime los codones de cada secuncia en formato FASTA. 

4. Imprime los codones de cada secuencia en los 6 marcos lectura.

<span style="color:green">El ejercicio se dividió en dos celdas, la primera responde los incisos 1 a 3 y la segunda el 4. El output de la segunda es el completo porque así lo pide la actividad.</span>

```python
import re
from Bio.Seq import Seq
from Bio import SeqIO

'''INPUT'''
path = 'seq.nt.fa'

# Gnero un diccionario que almacene los records del fasta
myDict = SeqIO.to_dict(SeqIO.parse(path, 'fasta'))

# Recorro el diccionario
for record in myDict.values():
    # Extraigo cada codón en una lista
    codones_list = re.findall('(.{3})', str(record.seq))
    # Una lista y la convierto en objeto Seq
    record.seq = Seq(' '.join(codones_list))
    # Imprimo la entrada
    print(record.format('fasta'))
```

```python
# Inicializo el directorio
myDict = SeqIO.to_dict(SeqIO.parse(path, 'fasta'))

# Recorro cada marco
# Los tres de foreard
for i in range(3):
    for record in myDict.values():
        # Obtengo codones con el dezplazamineto
        codones_list = re.findall('(.{3})', str(record.seq)[i:])
        # Genero el objeto Seq
        record.seq = Seq(' '.join(codones_list))
        # Imprimo el fasta con su marco
        print(f'Marco {i + 1} forward')
        print(record.format('fasta'))
    # Re inicializo el directorio para el siguiente marco
    myDict = SeqIO.to_dict(SeqIO.parse(path, 'fasta'))

# Los tres de reverse
for i in range(3):
    for record in myDict.values():
        # Obtengo codones con el dezplazamineto
        codones_list = re.findall('(.{3})', str(record.seq.reverse_complement())[i:])
        # Genero el objeto Seq
        record.seq = Seq(' '.join(codones_list))
        # Imprimo el fasta con su marco
        print(f'Marco {i + 1} reverse')
        print(record.format('fasta'))
    # Re inicializo el directorio para el siguiente marco
    myDict = SeqIO.to_dict(SeqIO.parse(path, 'fasta'))
```

## <span style="color:yellow"> Ejercicio_3</span>

<p>Obtén el <b>número</b> de lecturas cuyo promedio de calidad esté por debajo del umbral dado.</p>
<p><span class="given-return">Input:</span> Un umbral de calidad, Junto con un fastq.</p>


<span style="color:green">Aqui solo importé mean de statistics.</span>

```python
from Bio import SeqIO
from statistics import mean

'''INPUT'''
path = 'sample.fastq'
umbral = 38

# Inicializo un contador para los que estén debajo del umbral
coontador = 0
# Leo y recorro el fastq por cada record
for record in SeqIO.parse(path, 'fastq'):
    # Obtengo el promedio de la calidad de la lectura y lo comparo con el umbral
    if mean(record.letter_annotations["phred_quality"]) < umbral:
        # Aplico el contador
        coontador += 1

# Imprimo el resultado
print(coontador)
```

## <span style="color:yellow"> Ejercicio_5</span>
+ Consigue el *structure* del archivo pdb 4hhb

+ Imprime las coordenadas 3d del átomo C-alpha del aminoacido con residuo número 20 en chain C y asigna ese residuo a una variable.

+ Calcula el centro geométrico (centroid) del aminoacido mencionado anteriormente e imprímelo (esto asume que las masas de los átomos son iguales)

Para calular el centro geométrico se necesitan sacar las coordenadas de todos los átomos del residuo y sacar el promedio.

<span style="color:green">No hay mucho que decir, se describe con los comentarios.</span>

```python
from Bio import PDB

'''INPUT'''
path = '4hhb.pdb'

# Genero el parseador
parser = PDB.PDBParser(QUIET=True)

# Obtengo la structura
struc = parser.get_structure('prot_4hhb', path)

# Imprimo el objeto struc
print(f'La estructura: {struc}\n')

# Guardo el residuo y atomo objetivos
residue = struc[0]['C'][20]
calpha = residue['CA']

# Obtengo sus coordenadas
coord_capha = calpha.get_coord()
# Impriimo estas características obtenidas
print('Cordenadas del CA del residuo {} {}: {}\n'.format(residue.get_resname(), 
                                                       residue.get_id()[1], 
                                                       coord_capha))

# Para el centro geométrico inicializo tres int
x = 0
y = 0
z = 0
# Recorro cada atomo y sumo sus corrdenadas
for atom in residue:
    x += atom.get_coord()[0]
    y += atom.get_coord()[1]
    z += atom.get_coord()[2]

# Obtengo el centro geometrico a partir de calcular su media
natomos = len(residue)
centroid = [x/natomos, y/natomos, z/natomos]

# Imprimo los resultados
print('Centro geométrico de {} {} = {}'.format(residue.get_resname(), 
                                               residue.get_id()[1], 
                                               centroid))
```

## <span style="color:yellow"> Ejercicio_6</span>

1. Usando el archivo test.blastout.xml parsea los resultados e imprime lo siguiente "queryName	hitID	e-value	hitDesc	Hsp_num" Si el valor es mejor (meno) que 1e-10. (el query name lo encuentran en el blast_record y el hitDesc dentro del alignment.hit_def)

<span style="color:green">Se parsea el docuemnto y se imprime lo que pide la actividad, restringí el output a solo tres alineamientos para no saturar en exceso.</span>

```python
from Bio.Blast import NCBIXML

'''INPUT'''
path = 'test.blastout.xml'

texto = 'Query Name: {}\nHit ID: {}\ne-Value: {}\nHit Description: {}\nHsp Number: {}\n'

# Abro el archivo
with open(path, 'r') as filexml:
    # Lo parseo
    blast_records = NCBIXML.parse(filexml)
    # Recorro cada record
    for blast_record in blast_records:
        # Recorro cada alinieamineto
        for i, alignment in enumerate(blast_record.alignments):
            # Recorro cada característica hsp
            for hsp in alignment.hsps:
                
                # Si el e-value no es bueno se ignora
                if hsp.expect < 1e-10:
                    # Imprimo las carasterísticas
                    print(texto.format(
                        
                        blast_record.query_id,
                        alignment.hit_id,
                        hsp.expect,
                        alignment.hit_def, 
                        len(alignment.hsps)
                        
                    ))
            # Solo imprimo dee los tres primero alineamientos
            if i > 2:
                break
```
